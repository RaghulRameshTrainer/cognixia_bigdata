```python
#LOOP
#- While , For  ( Break and Continue)
```

```python
city=['Cochin','Chennai','Bangalore','Hyderabad','Pune','Delhi']

for x in city:
    if x == "Pune":
        break
    print(x)
```

```python
x=0
while x<=10:
    x = x + 1
    if x == 5:
        continue
    print(x)
```

```python

x=0
while x<=10:
    if x == 5:
        break
    print(x)
    x=x+1
print("I am out of loop")
```

```python
x=0

while x<=10:
    print(x)
    x=x+1
else:
    print("The value of x is : {} , Hence it came out of the loop!".format(x))
```
