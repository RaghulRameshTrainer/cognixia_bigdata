### 2. Join emails with user IDs
- Perform an outer join on **`convertedUsersDF`** and **`usersDF`** with the **`email`** field
- Filter for users where **`email`** is not null
- Fill null values in **`converted`** as **`False`**

Save result as **`conversionsDF`**.

```
conversionsDF=spark.sql("""SELECT u.user_id,u.email,u.user_first_touch_timestamp,
CASE WHEN u.email is not null THEN "true"
ELSE "false"
END AS coverted
 FROM users u LEFT OUTER JOIN convertedUsers c ON c.email=u.email 
       """)
display(conversionsDF)
```

### 3. Get cart item history for each user
- Explode **`items`** field in **`eventsDF`**
- Group by **`user_id`**
  - Collect set of all **`items.item_id`** objects for each user and alias with "cart"
  
Save result as **`cartsDF`**.

```
cartsDF=spark.sql("""
       SELECT user_id, collect_set(items['item_id']) as cart
       FROM (
      SELECT user_id,explode(items) as items FROM events
      ) tbl
      GROUP BY user_id
    """)
display(cartsDF)
```

### 4. Join cart item history with emails
- Perform a left join on **`conversionsDF`** and **`cartsDF`** on the **`user_id`** field

Save result as **`emailCartsDF`**.

```
conversionsDF.createOrReplaceTempView("conversions")
cartsDF.createOrReplaceTempView("carts")
emailCartsDF = spark.sql("""SELECT ca.user_id,email,coverted as converted FROM conversions co LEFT JOIN carts ca ON co.user_id=ca.user_id """)
display(emailCartsDF)
```

### 5. Filter for emails with abandoned cart items

### Bonus: Plot number of abandoned cart items by product

```
cartsDF=spark.sql(""" SELECT product, count(1) FROM (
       SELECT tbl.user_id, collect_set(items['item_name']) as product
       FROM (
      SELECT user_id,explode(items) as items FROM events
      ) tbl
       JOIN abandonedCarts ab ON tbl.user_id=ab.user_id
  GROUP BY tbl.user_id  ) t
  GROUP BY product
  ORDER BY count(1) DESC
    """)
display(cartsDF)
```
