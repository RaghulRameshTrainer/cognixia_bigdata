```
data=[["Raghul",38],["Bala",40],["Madhavan",45],["Malini",31]]
```

```
df=spark.sparkContext.parallelize(data).toDF(["Name","Age"])
df.show()
```

```
df.write.format("delta").mode("append").option("mergeSchema",True).saveAsTable("Customer")
```

```
spark.sql("SELECT * FROM Customer").show()
```

```
data1=[["Ramesh",38,"BE"],["Levin",40,"MCA"],["Saraswathi",45,"ME"],["Satya",31,"BSc"]]
rdd1=spark.sparkContext.parallelize(data1)
df1=rdd1.toDF(["Name","Age","Education"])
df1.show()
```

```
df1.write.format("delta").mode("append").option("mergeSchema",True).saveAsTable("Customer")
```

```
spark.sql("SELECT * FROM Customer").show()
```
